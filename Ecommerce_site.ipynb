#### DATA CLEANING

Will go through each column in the dataset and see what type of data is present in it. 
Whether any column or data needs to be altered for further analysis.

##### 1.  Brand

```python
df['Brand'].unique()
```

###### Note:
Since there are 8 unique brands in the dataset and there seem to be no discrepancy in this column including no null values so we leave it as it is.

##### 2. Form Factor

```python
df['Form_Factor'].unique()
```

###### Note:
Now, we can clearly see that there is 'Notebook' and 'Notebook, Compact' which is the same thing with an additional 'Compact' in the later. So, we will change all the 'Notebook, Compact' to 'Notebook'.
Also, there is 'Laptop' and 'Traditional Laptop'. Since all the Laptops are 'Clamshell' we will also change it to Laptop.

```python
df['Form_Factor'].replace(['Notebook, Compact','Notebook'],'NoteBook',inplace=True)
```

```python
df['Form_Factor'].replace(['Compact','Clamshell','Traditional Laptop'],'Laptop',inplace=True)
```

```python
df['Form_Factor'].unique()
```

##### 3. Item Weight

```python
df['Item_Weight']
```

##### Note:
Since the item weight is in kg and grams we will change the weight to its equivalent weight in 'kg' and store it in another column 'Item_Weight_in_Kg' and drop the original column 'Item_Weight'

```python
df['Item_Weight'].str.strip()
```

```python
df['Item_Weight']=df['Item_Weight'].apply(lambda w:w.split(' '))
```

```python
def weight(x):
    if len(x)>2:
        if((x[1].strip()=='kg') and (x[3].strip()=='g')):
            return x[0].strip()+'.'+x[2].strip()
    else:
        if(x[1].strip()=='kg'):
            return x[0].strip()
        else:
            return '0.'+x[0].strip()
    
```

```python
df['Item_Weight_in_Kg']=df['Item_Weight'].apply(weight)
```

```python
df.drop(['Item_Weight'],axis=1,inplace=True)
```

```python
df['Item_Weight_in_Kg']=pd.to_numeric(df['Item_Weight_in_Kg'],errors='coerce')
```

##### 5. Screen Size

```python
df['Screen_Size']
```

##### Note:
Since the screen size is available in 'Inches' as well as in 'Centimetres' so we will change all the size to its equivalent size in Inches 

```python
df['Screen_Size'].str.strip()
```

```python
df['Screen_Size']=df['Screen_Size'].apply(lambda s:str(s).split(' ') if " " in str(s) else str(s))
```

```python
def screen(x):
    if isinstance(x,list):
        if(x[1].strip()=='Inches'):
                return x[0].strip()
        else:
            if((x[1].strip()=='Centimetres') and (float(x[0])>20)):
                return (float(x[0])/2.54)
            else:
                return x[0]
    else:
        return x.strip()
    
```

```python
df['Screen_size_in_inches']=df['Screen_Size'].apply(screen)
```

```python
# Convert the datatype of the column to float with one decimal point values 
df['Screen_size_in_inches']=df['Screen_size_in_inches'].apply(lambda s:round(float(s),1))
```

```python
df.drop(['Screen_Size'],axis=1,inplace=True)
```

##### 6. Processor Brand

```python
df['Processor_Brand'].unique()
```

```python
df['Processor_Brand'].replace('?Intel','Intel',inplace=True)
```

##### 7. RAM Size

```python
df['Ram_size'].unique()
```

```python
df['Ram_size'].replace('?8 GB','8 GB',inplace=True)
df['Ram_size'].replace('?4 GB','4 GB',inplace=True)
```

```python
df['Ram_size'].str.strip()
```

```python
df['Ram_size']=df['Ram_size'].apply(lambda r:str(r).split(' ') if " " in str(r) else str(r))
```

```python
def ram(x):
    
    if isinstance(x,list):
        if(x[1]=='GB'):
            return x[0]
        elif(x[1]=='TB'):
            return (int(x[0])*1000)
    else:
        return x.strip()
    
```

```python
df['RAM_Size_GB']=df['Ram_size'].apply(ram)
```

```python
df.drop(['Ram_size'],axis=1,inplace=True)
```

```python
df['RAM_Size_GB']=pd.to_numeric(df['RAM_Size_GB'],errors='coerce')
```

##### 8. Battery Life

```python
df['Battery_Life'].unique()
```

```python
df['Battery_Life'].replace('?11 Hours','11 Hours',inplace=True)
df['Battery_Life'].replace('2 Years','2 Hours',inplace=True)
df['Battery_Life'].replace('3 Years','3 Hours',inplace=True)
```

```python
df['Battery_Life'].str.strip()
```

```python
df['Battery_Life']=df['Battery_Life'].apply(lambda b:str(b).split(' ') if " " in str(b) else str(b))
```

```python
def battery(x):
    if isinstance(x,list):
        if(x[1]=='Hours'):
            return x[0]
    else:
        return x
```

```python
df['Battery_Life']=df['Battery_Life'].apply(battery)
```

```python
df['Battery_Life']=pd.to_numeric(df['Battery_Life'],errors='coerce')
```

##### 9. Original Price

```python
df['Original_Price']
```

```python
df['Original_Price']=df['Original_Price'].apply(lambda x:str(x).replace(',','') if ',' in str(x) else str(x))
df['Original_Price']=df['Original_Price'].apply(lambda x:str(x).split('.')[0] if '.' in str(x) else str(x))
```

```python
df['Original_Price']=df['Original_Price'].apply(lambda x:str(x).lstrip())
```

```python
df['Original_Price']=pd.to_numeric(df['Original_Price'],errors='coerce')
```

```python
df['Original_Price'].round()
```

```python
df['Original_Price'].plot.hist()
```

```python
def inpute_mean(series):
    return series.fillna(series.mean())
```

```python
df['Original_Price']=df['Original_Price'].transform(inpute_mean)
```

```python
#df['Original_Price']=pd.to_numeric(df['Original_Price'],errors='coerce',downcast='float')
```

```python
df['Original_Price']=df['Original_Price'].round()
```

```python
df['Original_Price'].isnull().sum()
```

##### 10. Discounted Price

```python
df['Discount_Price']
```

```python
df['Discount_Price']=df['Discount_Price'].apply(lambda x:str(x).replace(',','') if ',' in str(x) else str(x)).astype(float)
```

```python
df['Discount_Price']=df['Discount_Price'].apply(lambda x:str(x).split('.')[0] if '.' in str(x) else str(x)).astype(float)
```

```python
df['Discount_Price']=pd.to_numeric(df['Discount_Price'],errors='coerce')
```

```python
df['Discount_Price'].round(1)
```

```python
df['Discount_Price'].plot.hist()
```

```python
df['Discount_Price']=df['Discount_Price'].transform(inpute_mean)
```

```python
#df['Discount_Price']=pd.to_numeric(df['Discount_Price'],errors='coerce')
```

```python
df['Discount_Price']=df['Discount_Price'].round()
```

```python
df['Discount_Price'].isnull().sum()
```

##### 11. Ratings

```python
df['Ratings']
```

```python
df['Ratings'].str.strip()
```

```python
df['Ratings']=df['Ratings'].apply(lambda ra:str(ra).split(' ') if " " in str(ra) else str(ra))
```

```python
def rating(x):
    if isinstance(x,list):
            return x[0]
    else:
        return x
```

```python
df['Ratings']=df['Ratings'].apply(rating)
```

```python
df['Ratings']=pd.to_numeric(df['Ratings'],errors='coerce')
```

```python
df['Ratings'].plot.hist()
```

```python
df['Ratings'].fillna(str(df['Ratings'].median()), inplace=True)
```

```python
df['Ratings']=pd.to_numeric(df['Ratings'],errors='coerce')
```

```python
df['Ratings'].isnull().sum()
```

```python
df['Discount']=df['Original_Price']-df['Discount_Price']
```

```python
def rep(x):
    if x <=40000:
        return "Low"
    if x>40000 and x<=60000:
        return 'Moderate'
    else:
        return 'High'
```

```python
df['Price_Range']=df['Discount_Price'].apply(rep)
```

```python
df.isnull().sum()
```
