```python
cols=data.columns
```

```python
data['Timestamp']=data['Timestamp'].str.split(' ')
```

```python
def time_d(x):
    y=x[0].split('/')
    for i in y:
        date=y[2]
        return date
```

```python
data['Date']=data['Timestamp'].apply(time_d)
```

```python
def time_m(x):
    y=x[0].split('/')
    for i in y:
        month=y[1]
        return month
```

```python
data['Month']=data['Timestamp'].apply(time_m)
```

```python
def time_y(x):
    y=x[0].split('/')
    for i in y:
        year=y[0]
        return year
```

```python
data['Year']=data['Timestamp'].apply(time_y)
```
